```python
import numpy as np
# Variance
data = np.array([1.2, 2.4, 3.5, 1.8, 4.0, 2.9])
# Calculating Mean
mean_val = np.mean(data)
# Calculating Variance
# Formula: sigma^2 = sum(x - mean)^2 / N
variance_val = np.var(data)
print(f"Dataset: {data}")
print(f"Mean: {mean_val:.2f}")
print(f"Variance: {variance_val:.2f}")
```

```python
# Two features (e.g., Study Hours vs Exam Scores)
study_hours = [2, 4, 6, 8, 10]
scores = [50, 60, 75, 85, 95]
# Calculate Covariance Matrix
cov_matrix = np.cov(study_hours, scores)
print("Covariance Matrix:")
print(cov_matrix)
print(f"\nCovariance between Hours and Scores: {cov_matrix[0, 1]:.2f}")
```

```python
A = np.array([[4, 2],
              [1, 3]])
# Calculate Eigenvalues and Eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)
print("Matrix A:")
print(A)
print("\nEigenvalues:")
print(eigenvalues)
print("\nEigenvectors (Columns):")
print(eigenvectors)
```

```python
# Standars Deviation
data_outliers = np.array([10, 12, 11, 13, 12, 100])

std_dev = np.std(data_outliers)
median_val = np.median(data_outliers)

print(f"Data with outlier: {data_outliers}")
print(f"Median (Robust to outliers): {median_val}")
print(f"Standard Deviation: {std_dev:.2f}")
```

```python
# Correlation between two variables
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10] # Perfectly linear

correlation_matrix = np.corrcoef(x, y)
r_value = correlation_matrix[0, 1]

print(f"Correlation Coefficient: {r_value}") # 1.0 means perfect positive correlation
```
