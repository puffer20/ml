```python
import numpy as np
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

def sigmoid_derivative(x):
    return x*(1-x)
```

```python
X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])

y = np.array([[0],
              [1],
              [1],
              [0]])
```

```python
#  Input layer → Hidden layer 1 → Hidden layer 2 → Output layer
#      2              4               3                1
```

```python
np.random.seed(1)

W1 = np.random.rand(2,4)   # input → hidden1
b1 = np.random.rand(1,4)

W2 = np.random.rand(4,3)   # hidden1 → hidden2
b2 = np.random.rand(1,3)

W3 = np.random.rand(3,1)   # hidden2 → output
b3 = np.random.rand(1,1)

learning_rate = 0.1
epochs = 20000
```

```python
for epoch in range(epochs):

    # ---------- Step 2: Forward Propagation ----------
    h1 = sigmoid(np.dot(X, W1) + b1)
    h2 = sigmoid(np.dot(h1, W2) + b2)
    y_pred = sigmoid(np.dot(h2, W3) + b3)

    # ---------- Step 3: Compute Loss (MSE) ----------
    loss = np.mean((y - y_pred)**2)

    # ---------- Step 4: Backpropagation ----------
    error_output = y - y_pred
    d_output = error_output * sigmoid_derivative(y_pred)

    error_h2 = np.dot(d_output, W3.T)
    d_h2 = error_h2 * sigmoid_derivative(h2)

    error_h1 = np.dot(d_h2, W2.T)
    d_h1 = error_h1 * sigmoid_derivative(h1)

    # ---------- Weight Updates ----------
    W3 += np.dot(h2.T, d_output) * learning_rate
    b3 += np.sum(d_output, axis=0)* learning_rate

    W2 += np.dot(h1.T, d_h2) * learning_rate
    b2 += np.sum(d_h2, axis=0)* learning_rate

    W1 += np.dot(X.T, d_h1) * learning_rate
    b1 += np.sum(d_h1, axis=0)* learning_rate

    if epoch % 1000 == 0:
        print("Epoch:", epoch, "Loss:", loss)
```
