```python
# PCA Implementation from scratch (Following Experiment Steps)

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Step 0: Create Sample Dataset
data = {
    'Feature1': [2,4,6,8,10],
    'Feature2': [1,3,5,7,9],
    'Feature3': [5,7,9,11,13]
}

df = pd.DataFrame(data)

print("Original Dataset:\n")
print(df)
```

```python
# Step 1: Standardize the dataset
X = df.values
mean = np.mean(X, axis=0)
std = np.std(X, axis=0)

X_standardized = (X - mean) / std

print("\nStandardized Data:\n")
print(X_standardized)
print(pd.DataFrame(X_standardized))
```

```python
# Step 2: Compute Covariance Matrix
cov_matrix = np.cov(X_standardized.T)

print("\nCovariance Matrix:\n")
print(cov_matrix)
```

```python
# Step 3: Compute Eigenvalues and Eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

print("\nEigenvalues:\n", eigenvalues)
print("\nEigenvectors:\n", eigenvectors)
```

```python
# Step 4: Sort eigenvalues in descending order
sorted_index = np.argsort(eigenvalues)[::-1]

sorted_eigenvalues = eigenvalues[sorted_index]
sorted_eigenvectors = eigenvectors[:, sorted_index]

print("\nSorted Eigenvalues:\n", sorted_eigenvalues)
```

```python
# Step 5: Select top k eigenvectors
k = 2
principal_components = sorted_eigenvectors[:, 0:k]

print("\nTop Principal Components:\n")
print(principal_components)
```

```python
# Transform data into new PCA space
X_pca = np.dot(X_standardized, principal_components)

print("\nTransformed Data (PCA Result):\n")
print(X_pca)
```

```python

pca_df = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])

print("\nPCA DataFrame:\n")
print(pca_df)
```
