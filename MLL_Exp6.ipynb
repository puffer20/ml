```python
import numpy as np
```

```python
# Step 2: Create input arrays
x1 = np.array([0, 0, 1, 1])
x2 = np.array([0, 1, 0, 1])
```

```python
# Step 3: Target output (example: AND gate)
t = np.array([0, 0, 0, 1])
```

```python
# Step 5 & 6: User input for weights
w1 = float(input("Enter W1 Weight Value: "))
w2 = float(input("Enter W2 Weight Value: "))
```

```python
# Step 7: User input for threshold
T = float(input("Enter Threshold Value: "))
```

```python
# Step 8: Calculate yin
yin = w1 * x1 + w2 * x2
print("Yin:", yin)
```

```python
# Step 10: Initialize output array
y = np.zeros(len(yin))
```

```python
# Step 11: Apply threshold using loop
for i in range(len(yin)):
    if yin[i] >= T:
        y[i] = 1
    else:
        y[i] = 0
```

```python
# Step 13 & 14: Print outputs
print("\nTarget O/P:", t)
print("Calculated O/P:", y)
```

```python
# Step 16: Compare arrays
if np.array_equal(y, t):
    print("Correct Weight And Threshold Values")
else:
    print("Incorrect Weights, Re-run Code")
```
