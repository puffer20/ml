```python
import numpy as np
```

```python
# Training dataset (Example: NAND gate)
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])
Y = np.array([1,1,1,0])   # Target outputs
```

```python
# Learning rate
n = 0.1

# Number of epochs
epochs = 10
```

```python
# Step 1: Initialize weights and bias
W = np.random.rand(2)
b = 0
print("Initial Weights:", W)
print("Initial Bias:", b)
```

```python
# Training Process
for epoch in range(epochs):
    print("\nEpoch:", epoch+1)

    for i in range(len(X)):

        # Step 2: Compute weighted sum
        z = np.dot(W, X[i]) + b

        # Step 3: Step activation function
        if z >= 0:
            y_hat = 1
        else:
            y_hat = 0

        # Step 4: Compute error
        e = Y[i] - y_hat

        # Step 5: Update weights and bias
        W = W + n * e * X[i]
        b = b + n * e

    # Display weights after each epoch
    print("Weights:", W)
    print("Bias:", b)
```

```python
print("\nFinal Weights:", W)
print("Final Bias:", b)
```

```python
# Testing
print("\nTesting Model:")
for i in range(len(X)):
    z = np.dot(W, X[i]) + b
    y_hat = 1 if z >= 0 else 0
    print(X[i], "->", y_hat)
```
